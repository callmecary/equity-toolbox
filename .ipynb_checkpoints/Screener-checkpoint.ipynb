```python
import quandl, json, pandas,datetime
from stockstats import StockDataFrame
CLIENT_SECRETS = 'client_secrets.json'
client_secrets = json.load(open(CLIENT_SECRETS))
quandl.ApiConfig.api_key = client_secrets["production"]["api_key"]
start_date = '2017-11-1'
end_date = pandas.to_datetime('today')
columns = ['Close']
#stock['macd']
#stock['macdh']
#stock['close_50_ema']
csv = pandas.read_csv('WIKI-datasets-codes.csv')
tickers = csv['Ticker']
for ticker in tickers:
    print('Calculating',ticker)
    df = quandl.get(ticker, start_date=start_date, end_date=end_date,collapse="daily")
    df_close = pandas.DataFrame(df, columns=columns)
    stock = StockDataFrame.retype(df_close)
    if(sum(stock['close_20_ema_xu_close_50_ema'].tail(3))>=1
      and stock['macdh'].tail(1).mean() >= 0
      ):
        print(ticker)
```
